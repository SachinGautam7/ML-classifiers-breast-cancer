```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load dataset
data = load_breast_cancer()
```

```python
label_names = data['target_names']
labels = data['target']
feature_names = data['feature_names']
features = data['data']
```

```python
print(label_names)
print('Class label = ', labels[0])
print(feature_names)
print(features[0])
```

```python
train, test, train_labels, test_labels = train_test_split(features, labels, test_size=0.30, random_state=42)
```

# Gaussian Naive Bayes

```python
# Initialize classifier
gnb = GaussianNB()
```

```python
# Train classifier
model = gnb.fit(train, train_labels)
```

```python
# Make predictions
preds = gnb.predict(test)
print(preds)
```

```python
# Evaluate accuracy
gnb_acc=accuracy_score(test_labels, preds)
gnb_acc
```

```python
print(classification_report(test_labels, preds))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=28)
```

```python
knn.fit(train, train_labels)
```

```python
knn_pred = knn.predict(test)
print(knn_pred)
```

```python
knn_acc=accuracy_score(test_labels, knn_pred)
knn_acc
```

```python
print(classification_report(test_labels, knn_pred))
```

# SVM

```python
from sklearn.svm import SVC
svm = SVC(C=3.0, kernel='rbf', degree=6, cache_size=1024)
```

```python
svm.fit(train, train_labels)
```

```python
svm_pred = svm.predict(test)
print(svm_pred)
```

```python
svm_acc=accuracy_score(test_labels, svm_pred)
svm_acc
```

```python
print(classification_report(test_labels, svm_pred))
```

# Light GBM

```python
import lightgbm as lgb

d_train = lgb.Dataset(train, train_labels)
params={}
params['learning_rate']=0.03
params['boosting_type']='gbdt' # GradientBoostingDecisionTree
params['objective']='multiclass' # Multiclass target feature
params['metric']='multi_logloss' # Metric for multiclass
params['max_depth']=15
params['num_class']=6 # No.of unique values in the target class not inclusive of the end value
```

```python
clf = lgb.train(params, d_train, 100)
```

```python
lgb_predictions = clf.predict(test)
```

```python
import numpy as np
```

```python
print(lgb_predictions[1], np.argmax(lgb_predictions[1]))
```

```python
lgb_pred = np.argmax(lgb_predictions, axis=1)
lgb_pred.shape
```

```python
lgb_acc=accuracy_score(test_labels, lgb_pred)
lgb_acc
```

```python
print(classification_report(test_labels, lgb_pred))
```

# Comparison

```python
import matplotlib.pyplot as plt

models = ["Gaussian NB", "KNN", "SVM", "Light GBM"]
f1_scores = [gnb_acc, knn_acc, svm_acc, lgb_acc]
f1_scores = [round(score, 2) for score in f1_scores]

# Create a horizontal bar chart
plt.barh(models, f1_scores, color='lightblue', height=0.6, edgecolor='white')
plt.xlabel("F1 Score")
plt.ylabel("Machine Learning Model")
plt.title("Comparison of F1 Scores")

# Add data labels to the right of bars
for i, value in enumerate(f1_scores):
    plt.text(value + 0.01, i, str(value), va='center')

# Remove axes boundaries
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
# plt.gca().spines['bottom'].set_visible(False)
# plt.gca().spines['left'].set_visible(False)

# plt.tick_params(bottom=False, left=False)  # Remove ticks on the axes
plt.show()
```
